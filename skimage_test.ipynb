```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from skimage.filters import meijering, sato, frangi, hessian
from skimage.color import rgb2gray
import matplotlib.pyplot as plt


# Load an example image
image = data.retina()
image_gray = rgb2gray(image)

# Apply the different filters
meijering_result = meijering(image_gray, sigmas=range(1, 10, 2))
sato_result = sato(image_gray, sigmas=range(1, 10, 2))
frangi_result = frangi(image_gray, sigmas=range(1, 10, 2))
hessian_result = hessian(image_gray, sigmas=range(1, 10, 2))

# Display results
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

axes[0, 0].imshow(image)
axes[0, 0].set_title('Original Image')

axes[0, 1].imshow(image_gray, cmap='gray')
axes[0, 1].set_title('Grayscale')

axes[0, 2].imshow(meijering_result, cmap='jet')
axes[0, 2].set_title('Meijering')

axes[1, 0].imshow(sato_result, cmap='jet')
axes[1, 0].set_title('Sato')

axes[1, 1].imshow(frangi_result, cmap='jet')
axes[1, 1].set_title('Frangi')

axes[1, 2].imshow(hessian_result, cmap='jet')
axes[1, 2].set_title('Hessian')

for ax in axes.ravel():
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Testing different sigma values for the Meijering filter
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sigmas_small = [1, 2]
sigmas_medium = [3, 4, 5]
sigmas_large = [6, 7, 8]

axes[0].imshow(meijering(image_gray, sigmas=sigmas_small), cmap='jet')
axes[0].set_title('Small Sigmas (1-2)')

axes[1].imshow(meijering(image_gray, sigmas=sigmas_medium), cmap='jet')
axes[1].set_title('Medium Sigmas (3-5)')

axes[2].imshow(meijering(image_gray, sigmas=sigmas_large), cmap='jet')
axes[2].set_title('Large Sigmas (6-8)')

for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
